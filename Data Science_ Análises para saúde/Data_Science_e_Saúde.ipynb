## Taxa de letalidade

```python
def latest_by_country(data):
  return data.groupby("Country/Region").sum().iloc[:,-1]

display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

```python
# taxa de letalidade

sum_up = combined.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

# GrÃ¡ficos finais

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-1], recovered.sum()[3:], deaths.sum()[3:]],
                  index=["confirmed", "recovered", "deaths"])
grouped_per_day = grouped_per_day.T
grouped_per_day.head()
```

```python
grouped_per_day["letality_rate_1"] = (grouped_per_day["deaths"] / grouped_per_day["confirmed"])*100
grouped_per_day["letality_rate_2"] = (grouped_per_day["deaths"] / (grouped_per_day["deaths"] + grouped_per_day["recovered"]))*100

grouped_per_day.head()
```

```python
plt.figure(figsize=(10,5))
grouped_per_day['letality_rate_1'].plot()
plt.title('Mortes por casos confirmados (Mundial)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = grouped_per_day['letality_rate_1'][3:362].plot()
ax = grouped_per_day['letality_rate_2'][3:362].plot()
plt.legend(['Mortes / (Casos Confirmados)', 'Mortes / (Mortes + Recuperados)'])
plt.title('Letalidade pessimista e otimista do primeiro ano (Mundial)', fontsize=14)
plt.grid(linestyle='-')
ax.set_ylabel('% Letalidade', fontsize=12)
ax.set_xlabel('fonte: alura.com.br')
plt.show()
```
