## Fazendo as primeiras visualizações

```python
import pandas as pd
df = pd.read_csv('data/monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(14,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura ao longo dos anos')
```
